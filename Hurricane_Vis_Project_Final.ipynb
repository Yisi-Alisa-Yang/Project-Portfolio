<font face="Times New Roman" size=5> 2. **Data Summary** <font> <br>
<br>
<font face="Times New Roman" size=4> The databases we will be using are: <font> <br>
<br>
<font face="Times New Roman" size=4> &nbsp; &nbsp; &nbsp; 1. Decadal average: Economic damages from disasters as a share of GDP (*Our World in Data*, https://ourworldindata.org/grapher/decadal-economic-disasters-type?tab=table): This dataset has 14 columns, including “Entity” (the country name), “Code” (the country code), “Year” (decadal), and 11 other columns of total economic damages from 11 different natural disasters as a share of GDP (decadal), such as the economic damages of landslide, wildfire, fog, etc.. The ones we will be using are: <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Entity”: This column includes 236 countries in the world.<br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Year”: In decadal, spanning from 1900 to 2020. <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Total economic damages from natural disasters as a share of GDP - Extreme weather (decadal)”: As indicated by its name, this column includes the total economic damages from extreme weather (storm) as a share of GDP of that country in that decade. This is the column we will use specifically because it represents the total economic damages caused by storms instead of other natural disaster types.
 <font> <br>
 <br>
<font face="Times New Roman" size=4> &nbsp; &nbsp; &nbsp; 2. All Natural Disasters: 1900-2021 (*Kaggle*, https://www.kaggle.com/datasets/brsdincer/all-natural-disasters-19002021-%20eosdis/data): This dataset includes 45 columns, which detailed the occurrence of that specific disaster, such as its “Year”, “Disaster Group” (natural or unnatural), “Disaster Subgroup” (such as climatological, geophysical, meteorological, etc.), “Disaster Type” (such as drought, earthquake, volcanic activity, etc.), “Country”, “Region”, “Continent”, “Location”, “Origin”, “Latitude”, “Longitude”, and so on. The columns that we will mostly be using are: <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Year”: This data is reported from 1900 to 2021, with no intervals, which indicates any recorded occurrences of all different types of natural disasters happened from 1900 to 2021 over the world.<br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Disaster Type”: This column includes all the different types of disasters. We are particularly interested in the “Storm” type, since this is the type that includes tropical cyclones. <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Disaster Subtype”: This column includes all the different subtypes of storm disasters, such as tropical cyclone, convective storm, and extratropical storm. We are particularly interested in the tropical cyclone subtype. <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Total Deaths”: This column includes the total number of deaths recorded for each disaster occurrence.
 <font> <br>
 <br>
 <font face="Times New Roman" size=4> &nbsp; &nbsp; &nbsp; 3. Hurricanes and Typhoons: 1851-2015 (*Kaggle*, https://www.kaggle.com/datasets/noaa/hurricane-database): There are two datasets included in this website, one is the records of all occurrences of hurricanes and typhoons in the Pacific Ocean area, the other one is the ones in the Atlantic Ocean area. Both datasets contain the same 22 columns, including “ID”, “Name”, “Date”, “Time”, “Status”, “Latitude”, “Longitude”, “Maximum Wind”, etc.. The columns that we will use are: <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “ID”: The specific unique ID of each hurricane or typhoon. We will use this to count the total number of unique ID and thus get the total number of their occurrences from year to year. <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Date”: The date (presented as YYYYMMDD) of the hurricanes and typhoons happened. We will just use the YYYY since we only want to know the total occurrences of tropical cyclones happened from year to year. The Pacific dataset spans from 1949 to 2015, while the Atlantic dataset spans from 1851 to 2015.
 <font> <br>
 <br>
<font face="Times New Roman" size=4> &nbsp; &nbsp; &nbsp; 4. Climate Change: Earth Surface Temperature Data (*Kaggle*, https://www.kaggle.com/datasets/berkeleyearth/climate-change-earth-surface-temperature-data?select=GlobalTemperatures.csv): This dataset has 9 columns, including “dt”, “LandAverageTemperature”, “LandAverageTemperatureUncertainty”, “LandMaxTemperature”, etc.. The ones we will be using the most are:<br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “dt”: This represents the date of the temperature that was recorded. It spans from 1750 to 2015. For the purpose of our research, we will only use the data from 1850 to 2015. <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “LandAverageTemperature”: This is the recorded average land temperature in the global scale. It will reflect the degrees and changes of climate change, particularly global warming.
 <font> <br>
 <br>
 <font face="Times New Roman" size=4> &nbsp; &nbsp; &nbsp; 5. Extreme Poverty (*Kaggle*, https://www.kaggle.com/datasets/mathurinache/extreme-poverty): This data has 4 columns, including “Entity”,  “Code”,  “Year”, and “<b>&#36;</b>30.00 per day - share of population below poverty line”. We will use all of the columns:<br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Entity”: This represents the country name. It includes 183 countries in the world. <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Code”: This is the country code of each entity. <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Year”: This is the year when the data were recorded. It spans from 1981 to 2019 in this dataset. <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “<b>&#36;</b>30.00 per day - share of population below poverty line”: This column includes the portion of the country’s population that is below the poverty line of <b>&#36;</b>30/day.
 <font> <br>
 <br>
  <font face="Times New Roman" size=4> &nbsp; &nbsp; &nbsp; 6. Population, 10,000 BCE to 2021 (*Our World in Data*, https://ourworldindata.org/grapher/population?tab=table&time=earliest..2021): This data has two columns, including “Entity”, “Code”, “Year”, “Population (historical estimates)”. We will use all of the columns:<br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Entity”: This represents the country name. It includes 254 countries in the world (some countries were formed later such as around the 1950s). <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Code”: This is the country code of each entity. <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Year”: This is the year when the data were recorded / estimated. It spans from 10,000 BCE to 2021 in this dataset. <br>
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; - “Population (historical estimates)”: This represents the historical estimates of the population of these countries from 10,000 BCE to 2021.
 <font> <br>
